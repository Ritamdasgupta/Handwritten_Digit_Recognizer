```python
import numpy as np
import pandas as pd
import csv
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

The X_train array contains the pixel intensity values for each digit, and Y_train contains the corresponding digit. These values have been read from the csv file using pandas utility "read_csv" as shown. X_test is the test set on which we carry out inferences with our neural network.

```python
cols = pd.read_csv("train_digit.csv", nrows=1).columns
df=pd.read_csv("train_digit.csv",delimiter=",",usecols=cols[1:])
df.fillna(0, inplace=True)
X_train=df.to_numpy()
df1=pd.read_csv("test_digit.csv",delimiter=",")
df1.fillna(0,inplace=True)
X_test=df1.to_numpy()
df2=pd.read_csv("train_digit.csv",delimiter=",",usecols=["label"])
Y_train=df2.to_numpy()

```

set_seed has been initialized for stable results. The neural network is a 11-layer DNN, and we avoid computational approximations by using SparseCategoricalCrossentropy() as our loss functions with from_logits=True, and putting activation in our output layer as linear instead of the usual softmax implementation. Adam optimizer has been used and training has been carried out for 200 epochs.

```python
tf.random.set_seed(1234)
model = Sequential(
    [               
        tf.keras.Input(shape=(784,)),    #specify input size
        ### START CODE HERE ###
        Dense(units=784, activation="relu"),
        Dense(units=400, activation="relu"),
        Dense(units=400, activation="relu"),
        Dense(units=400, activation="relu"),
        Dense(units=192, activation="relu"),
        Dense(units=96, activation="relu"),
        Dense(units=48, activation="relu"),
        Dense(units=48, activation="relu"),
        Dense(units=25, activation="relu"),
        Dense(units=15, activation="relu"),
        Dense(units=10, activation="linear")
        
        
        ### END CODE HERE ### 
    ], name = "my_model" 
)                            
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.001),
)

model.fit(
    X_train,Y_train,
    epochs=200
)
```

Using csv module utilities to write inferences onto "result_digit.csv". The variable "prediction" contains the regression output value given by our trained neural network. This is then converted to the probability values using the softmax() function, and finally the argmax() function takes the probability array and returns the digit corresponding to the highest probability. This, yhat, is our predicted digit and we store it onto the csv file as shown.

```python

with open('result_digit.csv', 'w', newline='') as file:
    fieldnames = ['ImageId', 'Label']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(len(X_test)):
        prediction = model.predict(X_test[i].reshape(1,784))
        prediction_p = tf.nn.softmax(prediction)
        yhat = np.argmax(prediction_p)
        writer.writerow({'ImageId': i+1, 'Label': yhat})
    
```
